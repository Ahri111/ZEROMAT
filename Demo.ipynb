```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from tabpfn import TabPFNRegressor
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

data = pd.read_csv("data/demo.csv", index_col = 0)
data.iloc[25:40]
```

```python
y_column = 'bandgap'
X_columns = [col for col in data.columns if col != y_column]
X = data[X_columns]
y = data[y_column]

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.1,
    random_state=42
)

regressor = TabPFNRegressor(device="cuda") 
regressor.fit(X_train, y_train)
print("Predicting...")
y_pred = regressor.predict(X_test)

# ÏÑ±Îä• ÌèâÍ∞Ä
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"R¬≤ Score: {r2:.4f}")
print(f"MAE: {mae:.4f}")
print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")

plt.scatter(y_test, y_pred, alpha=0.7, s=30)
min_val, max_val = min(y_test.min(), y_pred.min()), max(y_test.max(), y_pred.max())
plt.plot([min_val, max_val], [min_val, max_val], 'r--', lw=2)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title(f'Actual vs Predicted\nR¬≤ = {r2_score(y_test, y_pred):.3f}')
plt.grid(True, alpha=0.3)
```
